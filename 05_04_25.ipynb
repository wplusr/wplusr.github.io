## Fetch stock data

### Subtask:
Download historical stock data for Starbucks, Disney, Apple, Coca-Cola, and Walmart.


**Reasoning**:
Download historical stock data for the specified tickers using yfinance and display the head and info of the resulting dataframe.



```python
import yfinance as yf
import pandas as pd

tickers = ['SBUX', 'DIS', 'AAPL', 'KO', 'WMT']
stock_data = yf.download(tickers, start='2020-07-01', end='2025-07-01')
display(stock_data.head())
stock_data.info()
```

## Analyze stock data

### Subtask:
Perform various analyses on the stock data, such as calculating descriptive statistics, analyzing trends, or comparing performance.


**Reasoning**:
Calculate the daily percentage change of the 'Close' price for each stock, and then calculate the mean and standard deviation of these changes. Store these statistics in a new DataFrame.



```python
daily_returns = stock_data['Close'].pct_change()
mean_daily_returns = daily_returns.mean()
std_daily_returns = daily_returns.std()

performance_summary = pd.DataFrame({
    'Mean Daily Return': mean_daily_returns,
    'Standard Deviation of Daily Return': std_daily_returns
})

display(performance_summary)
```

## Visualize stock data

### Subtask:
Visualize the closing prices and daily returns of the stocks.


**Reasoning**:
Create the line plots for closing prices and daily returns, and the scatter plot for mean daily return vs risk.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Plotting Closing Prices
plt.figure(figsize=(12, 6))
sns.lineplot(data=stock_data['Close'])
plt.title('Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()

# 2. Plotting Daily Returns
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_returns)
plt.title('Stock Daily Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.grid(True)
plt.show()

# 3. Plotting Mean Daily Return vs. Risk
plt.figure(figsize=(10, 6))
sns.scatterplot(data=performance_summary, x='Standard Deviation of Daily Return', y='Mean Daily Return')
plt.title('Mean Daily Return vs. Risk')
plt.xlabel('Standard Deviation of Daily Return (Risk)')
plt.ylabel('Mean Daily Return')
for i, ticker in enumerate(performance_summary.index):
    plt.annotate(ticker, (performance_summary['Standard Deviation of Daily Return'][i], performance_summary['Mean Daily Return'][i]))
plt.grid(True)
plt.show()
```

### Subtask: Calculate and visualize cumulative returns

Calculate the cumulative returns for each stock and visualize them over time.

**Reasoning**:
Calculate the cumulative returns to see how an initial investment would have grown over time for each stock. Visualize this as a line plot to compare the overall performance.

```python
cumulative_returns = (1 + daily_returns).cumprod() - 1

plt.figure(figsize=(12, 6))
sns.lineplot(data=cumulative_returns)
plt.title('Cumulative Stock Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.grid(True)
plt.show()
```

## Calculate risk-adjusted returns (sharpe ratio)

### Subtask:
Calculate the Sharpe Ratio for each stock to assess risk-adjusted performance.


**Reasoning**:
Calculate the risk-free rate, excess daily returns, average excess daily returns, and annualized Sharpe Ratios for each stock, then display the results in a DataFrame.



```python
risk_free_rate_annual = 0.01
risk_free_rate_daily = risk_free_rate_annual / 252

excess_daily_returns = daily_returns - risk_free_rate_daily
average_excess_daily_returns = excess_daily_returns.mean()

sharpe_ratios = (average_excess_daily_returns / std_daily_returns) * (252**0.5)

sharpe_ratio_summary = pd.DataFrame({
    'Annualized Sharpe Ratio': sharpe_ratios
})

display(sharpe_ratio_summary)
```

## Analyze volatility over time

### Subtask:
Calculate and visualize the rolling standard deviation of daily returns to see how volatility has changed for each stock.


**Reasoning**:
Calculate and plot the rolling standard deviation of daily returns to visualize volatility changes over time.



```python
rolling_std = daily_returns.rolling(window=252).std()

plt.figure(figsize=(12, 6))
sns.lineplot(data=rolling_std)
plt.title('Rolling Volatility of Stock Returns (252-day window)')
plt.xlabel('Date')
plt.ylabel('Rolling Standard Deviation')
plt.grid(True)
plt.show()
```

## Perform drawdown analysis

### Subtask:
Calculate and visualize the drawdowns for each stock to understand the magnitude and duration of price declines.


**Reasoning**:
Calculate the cumulative maximum closing price and the drawdown for each stock, then visualize the drawdowns over time.



```python
cumulative_max = stock_data['Close'].cummax()
drawdown = (cumulative_max - stock_data['Close']) / cumulative_max

plt.figure(figsize=(12, 6))
sns.lineplot(data=drawdown)
plt.title('Stock Drawdown Over Time')
plt.xlabel('Date')
plt.ylabel('Drawdown')
plt.grid(True)
plt.show()
```
